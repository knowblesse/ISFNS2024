```python
def remove_signal(read_signal, true_signal):
    fft_true = np.fft.fft(true_signal)
    fft_read = np.fft.fft(read_signal)
    eps = 1e-12
    _fft_deconvolved = np.real(np.fft.ifft(fft_read / (fft_true + eps)))
    midpoint = int(_fft_deconvolved.shape[0] / 2)
    fft_deconvolved = np.append(_fft_deconvolved[midpoint::], _fft_deconvolved[:midpoint])
    return fft_deconvolved
plt.plot(remove_signal(convolved_step_function, step_function))
```

```python
# PSF for kernel
def sinc(x):
    return np.sin(x) / x

def psf(r, a, wavelength):
    return a**2 * sinc( (np.pi*a) / wavelength * np.sin(r) ) **2

def create_psf_kernel_2d(size, wavelength, a):
    # Create coordinate grids centered at (0, 0)
    x = np.linspace(-size // 2, size // 2, size)
    y = np.linspace(-size // 2, size // 2, size)
    x, y = np.meshgrid(x, y)

    # Convert Cartesian coordinates (x, y) to polar coordinates (r, theta)
    r = np.sqrt(x**2 + y**2)

    # Calculate the PSF
    psf_values = psf(r, a, wavelength)
    
    # Normalize the PSF so that its sum is 1
    psf_values /= np.sum(psf_values)
    
    return psf_values

psf_kernel_2d = create_psf_kernel_2d(56, 100, 1.5) * 256
plt.imshow(psf_kernel_2d, cmap='gray')
```
