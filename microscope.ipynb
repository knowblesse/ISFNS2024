```python
path_br = Path('/home/knowblesse/Data/Green/gCamp8F_mouse2_streaming_fps110_new')
#path_br = Path('/home/knowblesse/Data/ZSeries-07302024-0909-014')
paths = sorted(list(path_br.glob('*.tif')))
len(paths)
    
val = tiff.imread(paths[0])
plt.imshow(val[5], cmap='gray')
plt.scatter(70, 23, 10, 'r')
```

```python
from scipy.stats import zscore
sig = zscore(np.mean(np.mean(val[:, 20:26, 67:70], axis=1), axis=1))
sig_filt = np.convolve(sig, np.ones(10)/10, mode='valid')
plt.figure(figsize=(14, 8))
plt.plot(sig_filt)
vals = np.where(sig_filt > 0.8)[0]
plt.scatter(vals, sig_filt[vals], 60, 'r')
plt.xticks(np.arange(0, 2000, 200))
plt.xlim([400, 520])
plt.grid('on')
plt.ylim([-1, 3])
```
