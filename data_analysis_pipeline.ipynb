# Data analysis pipeline


```python
import numpy as np
import h5py
from pathlib import Path
import tifffile as tf
```

```python
# Load data
data_path = Path('/home/knowblesse/Data/TSeries-07302024-0909-018')
```

```python
tiffs = sorted(list(data_path.glob('*.tif')))
```

```python
tiffs
```

```python
#create in ctzyx
img = tf.imread(tiffs[0])
img = np.expand_dims(np.expand_dims(img, axis=1), axis=0)
empty_img = np.zeros_like(img)
img = np.concatenate((empty_img, img), axis=0).astype(np.int16)
print(img.shape)
print(img.dtype)
```

```python
# Write to h5 file
h5_path = data_path / 'data.h5'
with h5py.File(h5_path, 'w') as f:
    f.create_dataset('data', data=img)
    
```
